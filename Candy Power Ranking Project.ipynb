# Candy Power Ranking Project 

This project which candy is the most popular and will win in the ranking against other candies.  

Fivethirtyeight's ultimate Halloween candy power ranking dataset contains 85 different candies with their various characteristics.

Dataset is available at:
https://www.kaggle.com/fivethirtyeight/fivethirtyeight-candy-power-ranking-dataset

Description of the Attributes

'chocolate' - shows 1 if the candy contains chocolate or 0 if not.  chocolate.
'fruity' - shows 1 if the candy contains a fruit flavour or 0 if not. 
'caramel' - shows 1 if the candy contains caramel or 0 if not.
'peanutalmondy' - shows 1 if the candy contains peanuts, peanut butter or almonds or 0 if not.
'nougat' - shows 1 if the candy contains nougat or 0 if not.
'crispedricewafer' - shows 1 if the candy contains crisped rice, wafers, or a cookie componentis or 0 if not.
'hard' - shows 1 if the candy is hard or 0 if not.
'bar' - shows 1 if the candy is a candy bar or 0 if not.
'pluribus' - shows 1 if the candy is one of many candies in a bag or box or 0 if not.
'sugarpercent' - represents the percentile of sugar it falls under within the data set.
'pricepercent' - represents the unit price percentile compared to the rest of the set.
'winpercent' - represents the overall win percentage according to 269,000 matchups from 0 to 100.



### Explore the Model

This 'candy-data.csv' includes attributes for each candy along with its ranking. For binary variables, 1 means yes, 0 means no.The column win-percentage contains values between 0 and 100 specifying how often that candy was selected against with other candies.

```python
#load pandas
import pandas as pd

#read csv 
candy = pd.read_csv('candy-data.csv')

#view first 5 lines of data
candy.head()
```

### Split the Data

We will select the columns of chocolate, fruity, caramel, peanutyalmondy, nougat, crispedricewafer, hard, and bar to create the model. with the target variable being winpercent. This is shown with X below with a copy made of the selected columns and the winpercent column selected in y.

The dataset will then be split to create a model based on 80% of candies using the dataset X_train and need to report how accurate the model is at predicting the popularity of the 80% of candies the model was built on, and the 20%  of candies (X_test) it has never seen. 

```python
#import train_test_split
from sklearn.model_selection import train_test_split

#select the columns 
X = pd.get_dummies(candy.iloc[:,1:9])

#select the winpercent column
y = candy.iloc[:, 12]

#split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Create the Model

We will use the Support Vector Regression (SVR) because is that it is a non-parametric technique and the output model does not depend on distributions of the underlying dependent and independent variable. This is created below.

```python
# import SVR 
from sklearn.svm import SVR

# initiate the model
model = SVR()

# model is fit using X_train and y_train
model.fit(X_train, y_train)

# create vectors of predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

### Testing the Model

The Mean Absolute Error(MAE), mae(), can be used as the accuracy metric for the model. We can apply it to the test set and the test predictions. 

```python
# import mean_absolute_error
from sklearn.metrics import mean_absolute_error

# calculate the MAE
mae= mean_absolute_error(y_test, test_predictions
                         
# print MAE
print('The Mean Absolute Error is {}'.format(mae))
```

We can see that MAE is 10.28. The model can also be measured with a score shown below.

```python
# score the model
model.score(X_test, y_test)
```

This has a score of 17.5% this is a low score. This suggests these attributes have a 17.5% accuracy in predicting the winpercent of the candy.
